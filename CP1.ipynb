# Exploration of CompCars dataset

```python
import scipy.io
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_rows', None)
```

# Importing .mat files into python

```python
model_directory = 'data/misc/make_model_name.mat'
mat = scipy.io.loadmat(model_directory)
mat.items()
```

```python
mat['make_names']
```

```python
mat['model_names']
```

# Using dataframes to visualise the tables of model and make

```python
model_dataframe = pd.DataFrame(mat.get('model_names'))
model_dataframe.index = np.arange(1, len(model_dataframe) + 1)
model_dataframe.columns = ['model_name']
model_dataframe
```

```python
make_dataframe = pd.DataFrame(mat.get('make_names'))
make_dataframe.index = np.arange(1, len(make_dataframe) + 1)
make_dataframe.columns = ['make_names']
make_dataframe
```
